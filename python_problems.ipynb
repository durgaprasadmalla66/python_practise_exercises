```python
#guess the word game
word = "python"
guess = ""
guess_count = 0
guess_limit = 5
out_of_guesses = False
while guess != word and not out_of_guesses:
    if guess_count < guess_limit:
        guess = input("Enter your guess: ").lower()
        guess_count += 1
    else:
        out_of_guesses = True
if out_of_guesses:
    print("Out of guesses, you lose!")
else:    print("You win!")        
```

```python
#find the highest number in a list
numbers = [3, 5, 7, 2, 8, 10, 4 ,17]
highest = numbers[0]
for n in numbers:
    if n > highest:
        highest = n
print(highest)
```

```python
#Write a Python function that accepts two integer numbers. If the product of the two numbers is less than or equal to 1000, return their product; otherwise, return their sum.
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))
if x * y <= 1000:
    print(x * y)
else:
    print(x + y)
```

```python
#Iterate through the first 10 numbers (0–9). In each iteration, print the current number, the previous number, and their sum.
print("Printing the current number and previous number and the sum in range(10)")
previous_number = 0
for i in range(10):
    sum = previous_number + i
    print(f"Current Number: {i} Previous Number: {previous_number} Sum: {sum}")
    previous_number = i
```
